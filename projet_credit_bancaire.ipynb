```python
#Ce projet vise à mettre en place un modèle de données pour prédire si une demande de pret sera soldée ou pas
#On commence par importer les bibliothèques necessaires
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Nous chargons par la suite la base de données
db_pret = pd.read_csv("credit.csv")
db_pret["default"].fillna(value=db_pret["default"].mode(), inplace=True)
db_pret
```

```python
# Séparation des données
X = db_pret.drop("default", axis=1) # toutes les colonnes sauf la dernière (default)
y = db_pret["default"] # dernière colonne
```

```python
# Création de l'ensemble d'entraînement et de l'ensemble de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
db_pret
```

```python
# Mise en place du modèle de régression logistique sur les données d'entraînement
clf = LogisticRegression()
clf.fit(X_train, y_train)
```

```python
# Faire des prédictions sur les données de test
y_pred = clf.predict(X_test)
```

```python
# Déterminons la précision du modèle
acc = accuracy_score(y_test, y_pred)
print("Précision : ", acc)
```

```python
# Utilisons le modèle pour prédire si un nouveau prêt peut être remboursé ou non
nouveau_pret = [[41,3,17,15,176,10.3,12.359392,5.00860]] # données d'un nouveau prêt
statut_nouveau_pret = clf.predict(nouveau_pret)
print("Statut du nouveau prêt : ", statut_nouveau_pret)
```

```python
if statut_nouveau_pret == 1:
    print("Le client ne pourra pas remborser ce prêt")
else:
    print("Le client pourra remborser ce prêt")
```
